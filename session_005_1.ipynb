```python
import pickle
import pandas as pd
# deserialize persian df using pickle
with open('persian_df.pickle', 'rb') as handle:
    df_persian = pickle.load(handle)

df_persian.head()
```

```python
# we want to calculate tf and idf for all the words in the persian df
# we will use the persian_df we created in the previous sessions
# first we create a list of all the words in the persian_df
all_tokens_dict = {}

# first we iterate over each row
for index, row in df_persian.iterrows():
    cleaned_word_tokens = row["hazm_cleaned_word_token"]
    content = row["content"]
    cleaned_word_tokens_set = set(cleaned_word_tokens)
    for token in cleaned_word_tokens_set:
        if token not in all_tokens_dict:
            all_tokens_dict[token] = {}

all_tokens_dict
```

```python
import math
# now we get to the tf part
# we iterate over each row
for index, row in df_persian.iterrows():
    cleaned_word_tokens = row["hazm_cleaned_word_token"]
    description = row["content"]
    content = row["content"]
    content_to_list_of_words = content.split()
    cleaned_word_tokens_set = set(cleaned_word_tokens)
    for token in cleaned_word_tokens_set:
        # we check to see if the index is in the dict
        if index not in all_tokens_dict[token]:
            all_tokens_dict[token][index] = {}
        # we calculate the tf
        tf = cleaned_word_tokens.count(token) / len(content_to_list_of_words)
        all_tokens_dict[token][index]["tf"] = tf

# now we get to the idf part
# idf = log(N / df)
# N is the number of documents
# df is the number of documents that contain the word
# we iterate over each token
# first we get N which is number of rows in the persian_df
N = len(df_persian)
progress = 0
len_of_all_tokens_dict = len(all_tokens_dict)
for index, row in df_persian.iterrows():
    cleaned_word_tokens = row["hazm_cleaned_word_token"]
    cleaned_word_tokens_set = set(cleaned_word_tokens)
    for token in cleaned_word_tokens_set:
        if "df" not in all_tokens_dict[token]:
            all_tokens_dict[token]["df"] = 0
        all_tokens_dict[token]["df"] += 1

for token in all_tokens_dict:
    idf = math.log(N / all_tokens_dict[token]["df"])
    all_tokens_dict[token]["idf"] = idf

all_tokens_dict['کشیش']
```

```python
# now that we have the tf and idf for each word in each document
# we can calculate the tf-idf
# tf-idf = tf * idf
for token in all_tokens_dict:
    for index in all_tokens_dict[token]:
        if index != "df" and index != "idf":
            tf_idf = all_tokens_dict[token][index]["tf"] * all_tokens_dict[token]["idf"]
            all_tokens_dict[token][index]["tf_idf"] = tf_idf

all_tokens_dict['کشیش']
```

```python
# now we can have a simple query system that calculates the score for each document and returns the most relevant documents
sample_query = "کشیش به مدرسه رفت تا دانش آموزان را موعظه کند."
import hazm
normalizer = hazm.Normalizer()
tokenizer = hazm.WordTokenizer()
stopwords = hazm.stopwords_list()

def perform_search(query):
    # we need to calculate the score:
    # score = sum of tf-idf for each word in the query
    # we need to tokenize the query
    # we need to remove the stopwords
    # we need to normalize the query
    normalized_query = normalizer.normalize(query)
    tokenized_query = tokenizer.tokenize(normalized_query)
    query_without_stopwords = [word for word in tokenized_query if word not in stopwords]
    # now we iterate over each word in the query
    scores = {}
    for word in query_without_stopwords:
        if word in all_tokens_dict:
            for index in all_tokens_dict[word]:
                if index != "df" and index != "idf":
                    if index not in scores:
                        scores[index] = 0
                    scores[index] += all_tokens_dict[word][index]["tf_idf"]
    # now we have the scores for each document
    # we need to sort the scores
    sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    # now we have the sorted scores
    # we need to return the top 5 documents
    top_5 = sorted_scores[:5]
    # now we have the top 5 documents
    # we need to return the content of the top 5 documents
    top_5_content = []
    for index, score in top_5:
        top_5_content.append((df_persian.iloc[index]["content"], score))
    return top_5_content
    
perform_search(sample_query)
```

```python
# now we create a simple interface for the search engine
# we will use ipywidgets
from ipywidgets import interact

# return value is a list of strings 
# each string is the content of a document
# we will use the interact function from ipywidgets
# we will use the perform_search function we created
# we will use the query parameter
# we style the returned value using the HTML function
from IPython.display import HTML
def style_result(query):
    result = perform_search(query)
    html = ""
    for content in result:
        html += f"""
        <div style="border: 1px solid black; padding: 10px; margin: 10px;">
            score is {content[1]} and the content is <br> {content[0]}
        </div>
        """
    return HTML(html)

interact(style_result, query="")
```
